```python
from elasticsearch import Elasticsearch
import warnings
import csv
```

```python
warnings.filterwarnings("ignore")
```

```python
es = Elasticsearch(HOST="http://localhost", PORT=9200)
es = Elasticsearch()
```

```python
# Creating index
es.indices.create(index="translations")
#es.indices.delete(index="translations")
#es.indices.exists(index="translations")
```

```python
def filePath(language):
    return 'translations/'+language+'.csv'
```

```python
### If you want to run faster, choose this list instead of the original one
# languages = [ 'portuguese'] 

languages = ['deutsche', 'french', 'portuguese']

data = []
for language in languages:
    with open(filePath(language), newline='', encoding='utf-8') as csvfile:
        csvreader = csv.reader(csvfile, delimiter=',')
        for row in csvreader:
            englishSentence = row[0]
            sentence = row[1]
            data.append({ 'englishSentence': englishSentence, 'sentence': sentence, 'language': language })
            
print('TOTAL DATA LOADED: ' + str(len(data)))
```

```python
counter = 0

for q in data:
    print('INSERTING ROW ' + str(counter))
    es.index(index='translations', doc_type='sentence', id=counter, body=q)
    counter += 1
    
print('TOTAL DATA INSERTED INTO ELASTICSEARCH: ' + str(counter))
```

```python
# This will look for all english sentences that contain "Wait" and the possible translations in the language "portuguese"
body = {
    "from":0,
    "size":10,
    "query": {
        "bool": {
          "must": [
            {
              "match": {
                "language": "Portuguese",
              }
            },
            {
              "match": {
                "englishSentence": "Wait"
              }
            }
          ]
        }
    }
}

res = es.search(index='translations', body=body)

print(res)
```
